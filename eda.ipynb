```python
out=[]
for msg, grp in test_data.groupby('msg'):
    sim_max = max(grp.prob)
    if sim_max>0.5:
        max_index = grp.prob.idxmax()
        out.append(list(grp.loc[max_index])+[grp.out_node[max_index]])
    else:
        out.append(list(grp.loc[grp.index[0]])+['-99'])

out=pd.DataFrame(out)
```
